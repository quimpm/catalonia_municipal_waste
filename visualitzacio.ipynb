```python
df1 = df.dropna(subset=['Municipi', 'Comarca'])
df.info()
```

```python
df_no_moianes = df1[df1.Comarca != "Moianès"]
df_no_moianes
```

```python
import geopandas

def read_geofile(path_to_data):
    return geopandas.read_file(path_to_data)

def add_geometry_region_to_main_df(file, main_df):
    gdf = read_geofile(file).drop(columns = ["cap_comar", "comarca"])
    return (pd.merge(gdf, main_df, left_on='nom_comar', right_on='Comarca')
                .drop_duplicates(keep=False))

geo_df = add_geometry_region_to_main_df("./datasets/comarques-catalunya.geojson", df_no_moianes)
geo_df.info()
```
